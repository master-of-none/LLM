```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")

# Encode the input text
inputs = tokenizer.encode("Once upon a time ", return_tensors="pt")

# Generate text
outputs = model.generate(inputs, max_length=200, do_sample=True)

# Decode the generated sequence
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Tokenize the generated text
generated_tokens = tokenizer.tokenize(generated_text)

# Calculate perplexity using cross-entropy
with torch.no_grad():
    logits = model(inputs).logits
    target_ids = tokenizer.encode(generated_text, return_tensors="pt", max_length=200, truncation=True).to(logits.device)
    target_ids = target_ids[:, :logits.shape[1]]  # Ensure target batch size matches logits batch size
    loss = torch.nn.functional.cross_entropy(logits.view(-1, logits.size(-1)), target_ids.view(-1))
    perplexity = torch.exp(loss)

# Print the results
print(f"Generated Text:\n{generated_text}")
print(f"Total tokens: {len(generated_tokens)} \t Total unique tokens: {len(set(generated_tokens))}")
print(f"Type-Token Ratio: {len(set(generated_tokens)) / len(generated_tokens)}")
print(f"Perplexity: {perplexity.item()}")
```

```python
import matplotlib.pyplot as plt
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch


# Lists to store information for plotting
perplexity_values = []
token_counts = []
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloomz-1b7")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloomz-1b7")
inputs = tokenizer.encode("Once upon a time ", return_tensors="pt")

# Generate text multiple times and track perplexity
for _ in range(5):  # You can adjust the number of iterations
    inputs = tokenizer.encode("Once upon a time ", return_tensors="pt")
    outputs = model.generate(inputs, min_length=200, max_length=250, do_sample=True)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Calculate perplexity
    with torch.no_grad():
        input_ids = tokenizer.encode(generated_text, return_tensors="pt")
        logits = model(input_ids).logits
        labels = input_ids[:, 1:]
        loss = torch.nn.functional.cross_entropy(logits[:, :-1, :].reshape(-1, logits.shape[-1]), labels.reshape(-1), reduction="sum")
        perplexity = torch.exp(loss / labels.numel())

    # Append values for plotting
    perplexity_values.append(perplexity.item())
    token_counts.append(len(tokenizer.tokenize(generated_text)))

    #print(f"Generated Text:\n{generated_text}")
    #print(f"Perplexity: {perplexity.item()}")

# Plotting
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(perplexity_values, label='Perplexity')
plt.title('Perplexity over Generations')
plt.xlabel('Generation')
plt.ylabel('Perplexity')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(token_counts, label='Token Count')
plt.title('Token Count over Generations')
plt.xlabel('Generation')
plt.ylabel('Token Count')
plt.legend()

plt.tight_layout()
plt.show()
```
